```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
pd.pandas.set_option('display.max_columns', None)

sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

```python
# Read database 
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
# Summary of Database
df.info()
```

```python
# Checking for NULL values in the database
df.isnull().sum()
```

```python
# Summary of Database
df.describe()
```

```python
## Data Preprocessing 
```

```python
# Replacing null values in the column reviews_per_month with 0 in the dataset
df['reviews_per_month'].fillna(0,inplace = True)
```

```python
# Replacing null values in the column name with the character * and hostname with character @ in the dataset

df['name'].fillna("*",inplace=True)
df['host_name'].fillna("@",inplace=True)
```

```python
# Dropping the column last review as more than 10,000 data points contains null values.
df.drop(['last_review'],axis=1,inplace=True)
```

```python
## Distribution of Availability of rooms
```

```python
## Minimum nights people stay in different room types
```

```python
# In the private roooms people mostly stay for around 1-7 days depending upon the neighbourhood groups.
df1 = df[df.room_type=="Private room"]['minimum_nights']
f,ax = plt.subplots(figsize=(15,5))
ax = sns.swarmplot(y= df1.index,x= df1.values)
plt.xlabel("minimum_nights")
plt.show()
```

```python
#Mostly travellers, backpackers and people on low budget like to stay in the shared rooms.
#They live on an average of 1-2 days as they keep on moving from one place to another.
df1 = df[df.room_type=="Shared room"]['minimum_nights']
f,ax = plt.subplots(figsize=(15,5))
ax = sns.swarmplot(y= df1.index,x= df1.values)
plt.xlabel("minimum_nights")
plt.show()
```

```python
# On an average people live in these apartments from 1-90 days
df1 = df[df.room_type=="Entire home/apt"]['minimum_nights']
f,ax = plt.subplots(figsize=(15,5))
ax = sns.swarmplot(y= df1.index,x= df1.values)
plt.xlabel("minimum_nights")
plt.show()
```
